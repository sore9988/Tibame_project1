```python
import pandas as pd
train_df = pd.read_csv("中山區燒烤/中山區燒烤/train_all.csv", encoding="utf-8")
train_df
```

```python
train_dem_list = train_df["dem"].unique()
train_dem_list
```

```python
train_dd = train_df[["dem",  "star", "date", "text"]]
# 抓出不到一個月的 index => 改時間
date_2 = []
for l in train_dd["date"]:
    date_2.append("年" in l)
date2_index = train_dd[date_2].index
for dd in date2_index:
    train_dd.loc[dd,'date'] = '13 個月以上'
date_1 = []
for l in train_dd["date"]:
    date_1.append("天" in l)
date1_index = train_dd[date_1].index
for d in date1_index:
    train_dd.loc[d,'date'] = '1 個月內'
date_3 = []
for l in train_dd["date"]:
    date_3.append("小時" in l)
date3_index = train_dd[date_3].index
for d in date3_index:
    train_dd.loc[d,'date'] = '1 個月內'
train_dd.to_csv("中山區燒烤/中山區燒烤/train_dd_all.csv", encoding="utf-8", index=False)
train_dd
```

```python
# 依日期分
date_2_oo = [] # 一年前
date_2_o = [] # 半年前
date_2_n = [] # 半年內
for i in train_dd["date"]:
    date_2_oo.append(int(i[0:2]) >= 13)
    date_2_o.append(int(i[0:2]) > 6 and int(i[0:2]) <= 12)
    date_2_n.append(int(i[0:2]) <= 6)
train_oo = train_dd[date_2_oo]
train_o = train_dd[date_2_o]
train_n = train_dd[date_2_n]
train_oo
```

```python
trans = {v:k for k, v in enumerate(train_df["dem"].unique())}
reverse_trans = {k:v for k, v in enumerate(train_df["dem"].unique())}
trans
```

```python
# https://github.com/fxsjy/jieba
from urllib.request import urlretrieve
import os
import jieba
if not os.path.exists("dict.txt.big"):
    print("[1st time] 下載大辭典")
    url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
    urlretrieve(url, "dict.txt.big")
# with open("news.txt", "r", encoding="utf-8") as f:
#     news = f.read()
# 自備常用字典 及 可針對專科設定字典
jieba.set_dictionary("dict.txt.big")
jieba.load_userdict("politics_dict.txt")
# " ".join(jieba.cut(news))
```

```python
# apply重要 : data frame 的跟 series 的不同
def commendcut(s):
    s = " ".join(jieba.cut(s))
    s = s.replace("\r"," ").replace("\n"," ").replace('\" \"', "")
    return s
train = train_df["text"].apply(commendcut)  # 不要加 () 
```

```python
# 全部店家的關鍵字查詢
from jieba.analyse import extract_tags
train_g = ""
for i in train:
    train_g = train_g + i
catch_list = extract_tags(train_g, topK=2000)
s = ""
for i in train:
    s = s + " " + i
s = s.replace("評論 Google 翻譯", "")
fp = open("catch.txt", "w", encoding="utf-8")
fp.write(s)
fp.close()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x_train = vec.fit_transform(train)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train, y_train)
```

```python
# 輸入關鍵字查詢合適的店家
s = input("輸入關鍵字")
s = vec.transform([commendcut(s)])
pre = clf.predict(s)[0]
print(pre)
ans = reverse_trans[pre]
print("應該吃:", ans)
proba = clf.predict_proba(s)[0]
proba = list(enumerate(proba))
for i, prob in sorted(proba, key=lambda x:x[1], reverse=True):
    print(reverse_trans[i], ":", prob * 100, "%")
```
